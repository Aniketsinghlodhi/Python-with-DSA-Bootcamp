# Hollow Square of Side N

## Problem Description
Given an integer `n`, return a hollow square pattern of size n x n made up of `'*'`, represented as a list of strings.

- The hollow square has `'*'` on the border
- Spaces `' '` in the middle (except for side lengths of 1 and 2)

### Examples:
```
n = 1:       n = 2:       n = 4:
*            **           ****
             **           *  *
                          *  *
                          ****
```

````python
def hollow_square(n: int) -> list[str]:
    """
    Returns a hollow square pattern of size n x n.
    
    Args:
        n: The side length of the square
        
    Returns:
        A list of strings representing the hollow square pattern
    """
    if n <= 0:
        return []
    
    if n == 1:
        return ['*']
    
    if n == 2:
        return ['**', '**']
    
    result = []
    
    # First row: all stars
    result.append('*' * n)
    
    # Middle rows: star + spaces + star
    for _ in range(n - 2):
        result.append('*' + ' ' * (n - 2) + '*')
    
    # Last row: all stars
    result.append('*' * n)
    
    return result
````

````python
# Test the function with different values of n
for n in [1, 2, 3, 4, 5]:
    print(f"n = {n}:")
    pattern = hollow_square(n)
    for row in pattern:
        print(row)
    print()
````
