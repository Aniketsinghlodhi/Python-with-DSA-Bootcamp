# Hollow Square Pattern

## Problem Description
You are given an integer `n`. Your task is to return a hollow square pattern of size `n x n` made up of the character `'*'`, represented as a list of strings.

The hollow square has `'*'` on the border, and spaces `' '` in the middle (except for side lengths of 1 and 2).

### Example:
- **Input:** `n = 5`
- **Output:**
```
*****
*   *
*   *
*   *
*****
```

````python
def hollow_square(n: int) -> list[str]:
    """
    Generate a hollow square pattern of size n x n.
    
    Args:
        n: The side length of the square
        
    Returns:
        A list of strings representing each row of the hollow square
    """
    if n <= 0:
        return []
    
    if n == 1:
        return ['*']
    
    if n == 2:
        return ['**', '**']
    
    result = []
    
    # First row: all stars



    
    result.append('*' * n)
    
    # Middle rows: star, spaces, star
    for _ in range(n - 2):
        result.append('*' + ' ' * (n - 2) + '*')
    
    # Last row: all stars
    result.append('*' * n)
    
    return result
````

## Test Cases

````python
# Test with n = 5
n = 5
pattern = hollow_square(n)
print(f"Hollow Square (n={n}):")
for row in pattern:
    print(row)
````

````python
# Test with different values of n
for n in [1, 2, 3, 4, 6]:
    print(f"\nHollow Square (n={n}):")
    pattern = hollow_square(n)
    for row in pattern:
        print(row)
````
